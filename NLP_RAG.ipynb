### System and User Prompt Template

```python
# Define the system message — this sets the role and tone of your LLM
qna_system_message = (
    "You are a knowledgeable and reliable medical assistant. "
    "Your goal is to provide accurate, concise, and evidence-based medical information. "
    "Always explain your reasoning clearly, avoid assumptions, and do not provide diagnoses. "
    "If the context does not contain the answer, respond with 'The information is not available in the provided context.'"
)

# Define the user message template — this gives the context and the actual question
qna_user_message_template = (
    "Use the following context from a medical reference manual to answer the question. "
    "If the answer is not in the context, state that explicitly.\n\n"
    "Context:\n{context}\n\n"
    "Question: {question}\n\n"
    "Answer:"
)
```

### Response Function

```python
def generate_rag_response(user_input, k=10, max_tokens=128, temperature=0, top_p=0.95, top_k=50):
    global qna_system_message, qna_user_message_template

    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input, k=k)
    context_list = [d.page_content for d in relevant_document_chunks]

    # Combine all chunks into one context
    context_for_query = ". ".join(context_list)

    # --- Limit context to avoid exceeding model token window ---
    # Keep ~1800 tokens for input (leaving room for output)
    max_prompt_length = 1800
    if len(context_for_query) > max_prompt_length:
        context_for_query = context_for_query[:max_prompt_length]

    # Build user + system message
    user_message = qna_user_message_template.replace('{context}', context_for_query)
    user_message = user_message.replace('{question}', user_input)
    prompt = qna_system_message + '\n' + user_message

    # Generate the response
    try:
        response = llm(
            prompt=prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k
        )

        response = response['choices'][0]['text'].strip()
    except Exception as e:
        response = f"Sorry, I encountered the following error: \n{e}"

    return response
```

## Output Evaluation

Let us now use the LLM-as-a-judge method to check the quality of the RAG system on two parameters - retrieval and generation. We illustrate this evaluation based on the answeres generated to the question from the previous section.

- We are using the same Mistral model for evaluation, so basically here the llm is rating itself on how well he has performed in the task.

```python
# --- Define the system messages for evaluation ---

# Groundedness rater: checks if the answer is supported by retrieved context
groundedness_rater_system_message = """
You are a medical expert evaluating the groundedness of an AI-generated answer.
Given a question, the context retrieved from a medical document, and the model's answer,
rate how well the answer is supported by the provided context.
Return one of the following scores:
- 1: Poor grounding (answer not supported or contradicts context)
- 2: Partial grounding (somewhat supported, some missing or inconsistent info)
- 3: Fully grounded (fully supported by context, no contradictions)
Then provide a one-sentence justification for your score.
"""

```

```python
# Relevance rater: checks if the retrieved context and answer align with the query
relevance_rater_system_message = """
You are evaluating the relevance of a retrieved medical context and its generated answer
with respect to the user’s question. Assess whether the information retrieved and the
answer provided are relevant to the question.
Return one of the following scores:
- 1: Irrelevant (not related to the question)
- 2: Partially relevant (somewhat related, but incomplete or off-topic)
- 3: Highly relevant (directly and fully answers the question)
Then provide a one-sentence justification for your score.
"""

```

```python
# --- Template for question/context/answer ---
user_message_template = """
###Question
{question}

###Context
{context}

###Answer
{answer}
"""
```

```python
def generate_ground_relevance_response(user_input, k=3, max_tokens=128, temperature=0, top_p=0.95, top_k=50):
    global qna_system_message, qna_user_message_template

    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input, k=k)
    context_list = [d.page_content for d in relevant_document_chunks]

    # Combine and truncate context safely
    context_for_query = ". ".join(context_list)
    max_prompt_length = 1800   # Reserve ~500 tokens for model output
    if len(context_for_query) > max_prompt_length:
        context_for_query = context_for_query[:max_prompt_length]

    # Combine user prompt and system message
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""

    # Generate answer
    response = llm(
        prompt=prompt,
        max_tokens=max_tokens,
        temperature=temperature,
        top_p=top_p,
        top_k=top_k,
        stop=['INST'],
    )
    answer = response["choices"][0]["text"]

    # Build groundedness evaluation prompt
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Build relevance evaluation prompt
    relevance_prompt = f"""[INST]{relevance_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Generate both evaluation scores
    response_1 = llm(
        prompt=groundedness_prompt,
        max_tokens=max_tokens,
        temperature=temperature,
        top_p=top_p,
        top_k=top_k,
        stop=['INST'],
    )

    response_2 = llm(
        prompt=relevance_prompt,
        max_tokens=max_tokens,
        temperature=temperature,
        top_p=top_p,
        top_k=top_k,
        stop=['INST'],
    )

    return response_1['choices'][0]['text'], response_2['choices'][0]['text']
```

## Actionable Insights and Business Recommendations

**1. Broaden the dataset**
  * Some responses were only partially grounded because the system didn’t have enough coverage in certain topics. Expanding the document set with more domain-specific and detailed materials would strengthen future responses and improve factual grounding.

**2. Use stronger embeddings**
  * While the current embedding model performs well, using a more advanced or domain-adapted model, such as all-mpnet-base-v2 or BioClinicalBERT, could help capture subtle context and improve retrieval precision.

**3. Combine retrieval methods**
  * The evaluation revealed that certain answers drifted slightly off topic. A hybrid approach that combines semantic search with keyword-based retrieval (like BM25) could make responses more focused and contextually accurate.

**4. Optimize chunk sizes**
  * Some errors occurred because of long context windows or oversized chunks. Adjusting the chunk size dynamically depending on the content type can reduce overflow issues and improve retrieval quality.

**5. Maintain automated evaluation**
  * The LLM-as-a-judge evaluation method proved effective for assessing both groundedness and relevance. Keeping this evaluation mechanism active will provide continuous feedback on performance and guide future improvements.
